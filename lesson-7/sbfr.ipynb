```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('Solarize_Light2')

%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
data = pd.read_csv('opendata.csv', parse_dates=True, encoding='cp1251')
#data.columns
```

```python
print('Доступные категории: %s' % ', '.join(data['name'].unique()))
name = str(input('Введите категорию: '))
```

```python
print('Доступные регионы: %s' % ', '.join(data['region'].unique()))
region = 'Россия'
region = str(input(f'Введите регион (по умолчанию: {region}): '))
```

```python
min_date = data['date'].min()
max_date = data['date'].max()

print(f'Доступный интервал: {min_date} - {max_date}')
user_min = input(f'Введите дату начала (по умолчанию {min_date}): ')
user_max = input(f'Введите дату конца (по умолчанию {max_date}): ')

if user_min:
    min_date = user_min
    
if user_max:
    max_date = user_max
```

```python
selection = data.loc[(data['name'] == name) & (data['region'] == region) & (data['date'] >= min_date) & (data['date'] <= max_date)]
selection.head(10)
```

```python
plt.plot(selection['date'], selection['value'])
plt.title(' по '.join([name, region]), y=1.12)
plt.ylabel(name)
```
